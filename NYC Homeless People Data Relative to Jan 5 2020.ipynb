```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np 
NYCHomelessCounts= pd.read_csv('HOME-STAT_Weekly_Dashboard.csv')
NYCHomelessPeopleData = pd.read_csv('DHS_Daily_Report.csv')
```

```python
NYCHomelessCounts["CitywideCount"]
NYCHomessCountList = NYCHomelessCounts["CitywideCount"].tolist()
```

```python
def findDateIndex(series, column, Date):
    lst = series[column].tolist()
    L = len(lst)
    L1 = len(Date)
    currentIndex = 0
    for i in range(0, L):
        if lst[i][-L1:] == Date:
            currentIndex = i
            break
    return currentIndex
```

```python

JanIndex = findDateIndex(NYCHomelessCounts, "Week", "January 5th, 2020")
```

```python
def convertToPercentageIncreaseSinceDate(lst, dateIndex):
    L = len(lst)
    percentages = []
    dateValue = lst[dateIndex]
    for i in range(0,L):    
        percentage = 100*(lst[i]-dateValue)/dateValue
        percentages.append(percentage)
    return percentages
        
    
```

```python
NYCHomelessPercentages = convertToPercentageIncreaseSinceDate(NYCHomessCountList, JanIndex)
```

```python
x = NYCHomelessCounts["Week"][JanIndex:]
y = NYCHomelessPercentages[JanIndex:]
```

```python
plt.plot(x,y)
plt.xticks([0,10,15])
plt.xticks(fontsize=7)
plt.xlabel("Date")
plt.ylabel("Percentage Increase in Counts")
plt.title("Percentage Increase in Homeless Survey Counts Relative to first week of January 2020")
plt.show()
```
